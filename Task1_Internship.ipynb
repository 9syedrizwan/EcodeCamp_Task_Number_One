```python
#import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataset
df = pd.read_csv('Churn_Modelling.csv')
```

```python
# Check for missing values
print("Missing values:\n", df.isnull().sum())
```

```python
# Statistical summary of the dataset
print("\nSummary statistics:\n", df.describe())
```

```python
#Outlier Detection
plt.figure(figsize=(16, 8))
```

```python
# Plot for CreditScore
plt.subplot(2, 2, 1)
sns.boxplot(x='CreditScore', data=df, color='lightblue')
plt.title('Credit Score Box Plot')

# Plot for Age
plt.subplot(2, 2, 2)
sns.boxplot(x='Age', data=df, color='lightgreen')
plt.title('Age Box Plot')

# Plot for Balance
plt.subplot(2, 2, 3)
sns.boxplot(x='Balance', data=df, color='salmon')
plt.title('Balance Box Plot')

# Plot for EstimatedSalary
plt.subplot(2, 2, 4)
sns.boxplot(x='EstimatedSalary', data=df, color='orange')
plt.title('Estimated Salary Box Plot')

plt.tight_layout()
plt.show()
```

```python
# Exploratory Data Analysis
plt.figure(figsize=(16, 8))

# Histogram for CreditScore
plt.subplot(2, 2, 1)
sns.histplot(df['CreditScore'], kde=True, color='lightblue')
plt.title('Distribution of CreditScore')

# Histogram for Age
plt.subplot(2, 2, 2)
sns.histplot(df['Age'], kde=True, color='lightgreen')
plt.title('Distribution of Age')

# Histogram for Balance
plt.subplot(2, 2, 3)
sns.histplot(df['Balance'], kde=True, color='salmon')
plt.title('Distribution of Balance')

# Histogram for EstimatedSalary
plt.subplot(2, 2, 4)
sns.histplot(df['EstimatedSalary'], kde=True, color='orange')
plt.title('Distribution of Estimated Salary')

plt.tight_layout()
plt.show()
```

```python
#Correlation Matrix and Heatmap
plt.figure(figsize=(10, 6))
corr_matrix = df.corr(numeric_only=True) # Added numeric_only=True to only calculate correlation for numeric columns
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Visualize Relationships with Churn (target = 'Exited')
plt.figure(figsize=(12, 6))

# Churn rate by gender
plt.subplot(1, 2, 1)
sns.countplot(x='Gender', hue='Exited', data=df, palette='coolwarm')
plt.title('Churn by Gender')

# Churn rate by Geography
plt.subplot(1, 2, 2)
sns.countplot(x='Geography', hue='Exited', data=df, palette='coolwarm')
plt.title('Churn by Geography')

plt.tight_layout()
plt.show()
```

```python

# Scatter Plot to Visualize Churn vs. Numerical Features
plt.figure(figsize=(16, 8))

# Churn vs. CreditScore
plt.subplot(2, 2, 1)
sns.scatterplot(x='CreditScore', y='Exited', data=df, hue='Exited', palette='coolwarm')
plt.title('Churn vs. CreditScore')

# Churn vs. Age
plt.subplot(2, 2, 2)
sns.scatterplot(x='Age', y='Exited', data=df, hue='Exited', palette='coolwarm')
plt.title('Churn vs. Age')

# Churn vs. Balance
plt.subplot(2, 2, 3)
sns.scatterplot(x='Balance', y='Exited', data=df, hue='Exited', palette='coolwarm')
plt.title('Churn vs. Balance')

# Churn vs. EstimatedSalary
plt.subplot(2, 2, 4)
sns.scatterplot(x='EstimatedSalary', y='Exited', data=df, hue='Exited', palette='coolwarm')
plt.title('Churn vs. Estimated Salary')

plt.tight_layout()
plt.show()
```
